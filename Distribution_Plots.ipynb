```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
penguins = sns.load_dataset('penguins')
_ = sns.displot(data=penguins,x='flipper_length_mm')
```

```python
_ = sns.displot(data=penguins,x='flipper_length_mm',binwidth=3)
```

```python
tips=sns.load_dataset('tips')
_ = sns.displot(data=tips,x='size')
```

```python
_ = sns.displot(data=tips,x='size',discrete=True)
```

```python
_ = sns.displot(data=tips,x='size',shrink=1)
```

```python
_ = sns.displot(data=penguins,x='flipper_length_mm',hue='species')
```

```python
_ = sns.displot(data=penguins,x='flipper_length_mm',hue='species',element='step')
```
