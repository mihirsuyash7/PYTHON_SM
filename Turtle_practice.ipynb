```python
from turtle import *
```

```python
# Letter 'S'
pencolor('blue')
pensize(3)
speed(0) 
for i in range(160):
    backward(1)
    left(1)
for i in range(200):
    backward(1)
    right(1)    
mainloop()
```

```python
# Hour Glass 
color('red')
pensize(5)
speed('fast')
fd(100)
right(90)
penup()
fd(100)
right(90)
pendown()
fd(100)
right(90)
penup()
fd(100)
pendown()
pencolor('green')
right(135)
forward(100*1.414)
left(135)
penup()
fd(100)
left(135)
pendown()
forward(100*1.414)
right(45)
'''pencolor('blue')
for i in range(160):
    fd(1)
    right(1.15)
'''
mainloop()
```

```python
# Letter 'U'
pencolor('blue')
pensize(3)
speed('slow')
right(90)
fd(65)
for i in range(90):
    fd(1)
    left(2)
fd(65)
mainloop()
```
