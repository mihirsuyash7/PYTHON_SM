```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('watches.csv')
df.head(11)
```

```python
def clean_price_column(price):
    if isinstance(price,str):
        price=price.replace(',','').replace('₹','')
        return price
    return price
```

```python
def clean_rating_column(rating):
    if isinstance(rating,str):
        if((len(rating)==3) and '.' in rating):
            return float(rating)
        return np.nan
    return float(rating)
```

```python
df['New_Price']=df['price'].apply(clean_price_column)
```

```python
df['New_Rating']=df['rating'].apply(clean_rating_column)
```

```python
df['New_Oprice']=df['oprice'].apply(clean_price_column)
```
